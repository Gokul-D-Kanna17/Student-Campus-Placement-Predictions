```python
# import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading the file
dataset = pd.read_csv('Python data final.csv')
dataset
```

```python
# dropping the stream column
dataset = dataset.drop('Stream', axis=1)
dataset
```

```python
# catgorising col for further labelling
dataset["Current Degree"] = dataset["Current Degree"].astype('category')
dataset["Backlog"] = dataset["Backlog"].astype('category')
dataset[" Work Experience "] = dataset[" Work Experience "].astype('category')
dataset["Placement Status"] = dataset["Placement Status"].astype('category')
dataset.dtypes
```

```python
# labelling the columns
dataset["Current Degree"] = dataset["Current Degree"].cat.codes
dataset["Backlog"] = dataset["Backlog"].cat.codes
dataset[" Work Experience "] = dataset[" Work Experience "].cat.codes
dataset["Placement Status"] = dataset["Placement Status"].cat.codes
dataset
```

```python

# selecting the features and labels
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, -1].values
 
# display dependent variables
Y
X
```

```python

# dividing the data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
                                                    test_size=0.2)
 
# display dataset
dataset.head()
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel
cls= svm.SVC(kernel='poly') #polynomial kernel
clfs= svm.SVC(kernel='rbf') #Gaussian Radial Basis Function

#Train the model using the training sets
clf.fit(X_train, Y_train)
cls.fit(X_train, Y_train)
clfs.fit(X_train, Y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
y_pred1 = cls.predict(X_test)
y_pred2 = clfs.predict(X_test)

print(y_pred,y_pred1,y_pred2)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy for linear kernel:",metrics.accuracy_score(Y_test, y_pred))
print("Accuracy for polynomial kernel:",metrics.accuracy_score(Y_test, y_pred1))
print("Accuracy for rbf kernel:",metrics.accuracy_score(Y_test, y_pred2))

print(" ")
```

```python
# Model Precision: what percentage of positive tuples are labeled as such?

print("Precision for linear kernel:",metrics.precision_score(Y_test, y_pred,average='weighted'))
print("Precision for polynomial kernel:",metrics.precision_score(Y_test, y_pred1,average='weighted'))
print("Precision for rbf kernel:",metrics.precision_score(Y_test, y_pred2,average='weighted'))

print(" ")

# Model Recall: what percentage of positive tuples are labelled as such?
print("Recall for linear kernel:",metrics.recall_score(Y_test, y_pred,average='weighted'))
print("Recall for polynomial kernel:",metrics.recall_score(Y_test, y_pred1,average='weighted'))
print("Recall for rbf kernel:",metrics.recall_score(Y_test, y_pred2,average='weighted'))
```
